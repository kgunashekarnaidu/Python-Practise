<h3>A company wants to calculate the final salary of an employee after adding a bonus. The
bonus depends on the employee’s years of experience and monthly salary.<h3>

<h4>Bonus Rules:
    
• If the employee has more than 5 years of experience and the salary is less than
50,000/- : Add 10% bonus to the salary
    
• If the employee has 3 to 5 years of experience: Add 5% bonus to the salary

• If the employee has less than 3 years of experience: No bonus</h4>

```python
# Taking input from the users like Employee name ,Experience,Salary
Employee_name = input("Enter the Employee name :")
Salary = float(input("Enter the Monthly salary :" ))
Years_of_Experience = int(input("Enter the Years of Experience :"))
```

```python
#calculating the bonus amount for employee by creating a function and conditional statements
def calculated_bonus():
    if Years_of_Experience > 5 and Salary <50000.00:
        bonus = Salary * 0.1
    elif 3 >= Years_of_Experience <= 5:
        bonus = Salary * 0.05
    else:
        bonus = 0
    return bonus
        
```

```python
# storing the return values
Bonus_amount= calculated_bonus()
```

```python
#calculating the final salary of employee
Final_amount = Salary + Bonus_amount
```

```python
# Displaying the Ouput using Print()
print("FINAL EMPLOYEE SALARY")
print(f"EMPLOYEE NAME : {Employee_name}")
print(f"EXPERIENCE(in yrs) : {Years_of_Experience}")
print(f"BONUS AMOUNT : {Bonus_amount}")
print(f"FINAL SALARY : {Final_amount}")
```

<h3>A training institute wants to decide whether a student is eligible to receive a certificate.
The eligibility depends on the student’s attendance percentage and marks obtained in
subjects.</h3>

<h4>Eligibility Rules:</h4>
<h5>
A student is Eligible for Certification only if both conditions are satisfied:

    
• Attendance must be at least 75%

• Marks in all subjects must be 40 or above

• If any one of the above conditions is not satisfied, the student is Not Eligible.</h5>

```python
#Taking the input from the user
Student_name = input("Enter the students Name : ")
Total_classes = int(input("Enter the Total number of classes conducted : "))
Classes_attended = int(input("Enter number of classes attended:"))
```

```python
#calculating attendance
Attendance = (Classes_attended/Total_classes) * 100
```

```python
Attendance
```

```python
Marks = []
for i in range (1,4):
    mark = float(input("Enter the marks: "))
    Marks.append(mark)
```

```python

def checking_eligiblity():
    if Attendance > 75.0 and min(Marks) > 40:
        return "Eligible"
    else :
        return "Not Eligible"
```

```python
#storing theretuen values
Eligibility = checking_eligiblity()
```

```python
print(f"STUDENT NAME : {Student_name}")
print(f"ATTENDANCE : {Attendance}")
print(f"MARKS OF ALL SUBJECTS : {Marks}")
print(f"ELIGIBILITY STATUS: {Eligibility}")
```
