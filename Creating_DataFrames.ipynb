<h3>Creating the Data Frame using Lists</h3>

```python
# creating a list of data
lst_1 = [['D101','Dr.Anil Kumar','Cardiologist',12,'chennai'],
       ['D102','Dr. Priya Singh','Dermatologist',8,'Delhi'],
       ['D103','Dr.Ramesh','Neurologist',15,'Bengaluru'],
       ['D104','Dr.Nirinjan Kumar','Pediatrician',6,'Mumbai'],
       ['D105','Dr. suresh Rao','Orthopedic',10,'Hyderabad']]
```

<h3>Creating DataFrames using Dictionaries</h3>

```python
# Creating a Dictionary 
data = {
    'Doctor ID' : ('D101','D102','D103','D104','D105'),
    'Name':['Dr.Anil Kumar','Dr. Priya Singh','Dr.Ramesh','Dr.Nirinjan Kumar','Dr. suresh Rao'],
    'Specialization': ['Cardiologist','Dermatologist','Neurologist','Pediatrician','Orthopedic'],
    'Experience(Years)' :[12,8,15,6,10],
    'City' : ['chennai','Delhi', 'Bengaluru', 'Mumbai', 'Hyderabad']
}
```

```python
#Creating a DataFrames using Dictionary
df = pd.DataFrame(data, columns = cols)
df
```
